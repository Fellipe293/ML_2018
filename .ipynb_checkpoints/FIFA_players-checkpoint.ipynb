# Introdução - O Problema

O presente trabalho tem por objetivo estimar a posição preferida dos jogadores do jogo FIFA da EA Sports através do algoritmo K-Nearest Neighbors e em seguida comparar com o que é esperado pelos dados da FIFA.

# Base de Dados

## Input - X

Como dados de entrada utilizamos features dos jogadores fornecidas pelo Kaggle em https://www.kaggle.com/kevinmh/fifa-18-more-complete-player-dataset/version/1/data. Essas features são atributos de habilidades de cada jogador, todas na escala de 0 a 100. Features categóricas foram desconsideradas para o modelo em questão. Isso nos deixou essencialmente com dois tipos de variáveis: as consolidadas - como pace, passing, shot e drible, e as pulverizadas - como crossing, finishing, volleys e interceptions. Sendo assim, separamos nosso banco de dados em diferentes dataframes para rodar nosso modelo durante a validação. Após isso, escolhemos aquele que melhor se adaptou para implementação do modelo final.

## Target - Y

Como target criamos a coluna Position que indica a posição preferida pelo jogador. Ela foi criada por um critério de Voting utilizando as colunas prefersXX, onde XX indica as diferentes posições possíveis no jogo. Para simplificar a modelagem, reduzimos as diferentes posições a três grandes grupos: Ataque, Meio Campo e Defesa. Os casos de empate dos votos foram descartados para não acrescentar ruído na predição.

# Pre-processamento

Fizemos a normalização dos dados utilizando o StandardScaler do pacote sklearn inserido numa pipeline com o modelo KNN para serem aplicados em conjunto durante a validação cruzada.

Buscamos aplicar PCA também para realizar redução de dimensionalidade. Todavia, não houve muito ganho de tempo então acabamos descontinuando a ideia.

# Modelo

O modelo escolhido foi o K-Nearest Neighbors (KNN) pois é um modelo prático de ser implementado, além de possuir um treino rápido, apesar de ser lento para avaliar o erro de teste.

# Score

Utilizando os dados pulverizados, conseguimos alcançar uma acurácia de $94\%$, a qual se dá pela comparação do nosso modelo com o que era esperado de acordo com os resultados da FIFA.

```python
players_atributes = pd.read_csv('complete.csv', encoding='latin1')

variaveis_indesejadas = {
    'primeira_camada' : [
        'name','club','club_logo','flag','photo',
        'eur_value','eur_wage','release_clause_eur',
        'body_type','real_face','international_reputation',
        'league', 'special', 'rs', 'rw', 'rf', 'ram', 'rcm', 
        'rm', 'rdm', 'rcb', 'rb', 'rwb', 'st', 'lw', 'cf', 'cam',
        'cm', 'lm', 'cdm', 'cb', 'lb', 'lwb', 'ls', 'lf', 'lam',
        'lcm', 'ldm', 'lcb', 'gk_diving', 'gk_handling',
        'gk_kicking','gk_positioning','gk_reflexes', 'gk_flat_kick_trait',
        'gk_long_throw_trait', 'gk_up_for_corners_trait', 'prefers_gk', '1_on_1_rush_trait',
        'acrobatic_clearance_trait', 'argues_with_officials_trait','avoids_using_weaker_foot_trait',
        'backs_into_player_trait', 'bicycle_kicks_trait', 'cautious_with_crosses_trait',
        'chip_shot_trait', 'chipped_penalty_trait', 'comes_for_crosses_trait',
        'corner_specialist_trait', 'diver_trait', 'dives_into_tackles_trait',
        'diving_header_trait', 'driven_pass_trait', 'early_crosser_trait',
        "fan's_favourite_trait", 'fancy_flicks_trait', 'finesse_shot_trait',
        'flair_trait', 'flair_passes_trait', 'giant_throw_in_trait',
        'inflexible_trait', 'injury_free_trait', 'injury_prone_trait',
        'leadership_trait', 'long_passer_trait', 'long_shot_taker_trait', 
        'long_throw_in_trait', 'one_club_player_trait', 'outside_foot_shot_trait',
        'playmaker_trait', 'power_free_kick_trait', 'power_header_trait',
        'puncher_trait', 'rushes_out_of_goal_trait', 'saves_with_feet_trait',
        'second_wind_trait', 'selfish_trait', 'skilled_dribbling_trait',
        'stutter_penalty_trait', 'swerve_pass_trait', 'takes_finesse_free_kicks_trait',
        'target_forward_trait', 'team_player_trait', 'technical_dribbler_trait',
        'tries_to_beat_defensive_line_trait', 'poacher_speciality',
        'speedster_speciality', 'aerial_threat_speciality', 'dribbler_speciality',
        'playmaker_speciality', 'engine_speciality', 'distance_shooter_speciality',
        'crosser_speciality', 'free_kick_specialist_speciality', 'tackling_speciality',
        'tactician_speciality', 'acrobat_speciality', 'strength_speciality',
        'clinical_finisher_speciality'],
    
    'segunda_camada' : ['gk', 'prefers_st',
        'prefers_cf', 'prefers_lf', 'prefers_rf', 'prefers_rs',
        'prefers_ls', 'prefers_cdm', 'prefers_cm',
        'prefers_cam', 'prefers_lm', 'prefers_rm',
        'prefers_rw', 'prefers_lw', 'prefers_lam',
        'prefers_lcm', 'prefers_ldm', 'prefers_ram',
        'prefers_rcm', 'prefers_rdm', 'prefers_rb',
        'prefers_rwb', 'prefers_cb', 'prefers_lb',
        'prefers_lwb', 'prefers_rcb', 'prefers_lcb',
        'preferred_foot', 'work_rate_def', 'work_rate_att', 
        'weak_foot', 'ID', 'Votes_Attack', 'Votes_Middle',
        'Votes_Defense'],
    
    'terceira_camada' : ['birth_date','Sign']
}

class_group = {
    'Ataque':
        ['prefers_st',
        'prefers_cf',
        'prefers_lf',
        'prefers_rf',
        'prefers_rs',
        'prefers_ls'],
    'Meio_campo':
        ['prefers_cdm',
        'prefers_cm',
        'prefers_cam',
        'prefers_lm',
        'prefers_rm',
        'prefers_rw',
        'prefers_lw',
        'prefers_lam',
        'prefers_lcm',
        'prefers_ldm',
        'prefers_ram',
        'prefers_rcm',
        'prefers_rdm'],
    'Defesa':
        ['prefers_rb',
        'prefers_rwb',
        'prefers_cb',
        'prefers_lb',
        'prefers_lwb',
        'prefers_rcb',
        'prefers_lcb']
}


players_atributes['Votes_Attack'] = players_atributes[class_group['Ataque']].sum(axis=1)
players_atributes['Votes_Middle'] = players_atributes[class_group['Meio_campo']].sum(axis=1)
players_atributes['Votes_Defense'] = players_atributes[class_group['Defesa']].sum(axis=1)


lista = []
count = 0
def decide(a,b,c):
    m = max(a,b,c)
    options = []
    if a == m:
        options.append('Ataque')
    if b == m:
        options.append('Meio Campo')
    if c == m:
        options.append('Defesa')   
    if len(options) == 1:
        return options[0]
    else:
        global count 
        count = count + 1
        return "Deletar"

    
for i in players_atributes.index:
    lista.append(decide(players_atributes.loc[i,'Votes_Attack'],
                        players_atributes.loc[i,'Votes_Middle'],
                        players_atributes.loc[i,'Votes_Defense']))
    
players_atributes['Position'] = lista

players_atributes = players_atributes.drop(labels = variaveis_indesejadas['primeira_camada'], axis = 1)
players_atributes = players_atributes[players_atributes['Position'] != 'Deletar']
players_atributes = players_atributes.drop(labels = variaveis_indesejadas['segunda_camada'], axis = 1)
players_atributes[['overall','age', 'pac', 'sho','pas','dri','def',
                   'phy']].hist(figsize = (12,12), grid = False)
plt.show()
```

## PCA

```python
pca = PCA(n_components=25)
x_pca = players_atributes.drop(labels = 'Position',axis = 1)
y_pca = players_atributes['Position']

columns = x_pca.columns

scaler.fit(x_pca)

x_pca = scaler.transform(x_pca)  
principalComponents = pca.fit_transform(x_pca)

print(pca.explained_variance_ratio_)
sum(pca.explained_variance_ratio_)
```

```python
X = x_pca
y = y_pca
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)

cv_scores = []
k_options = range(11,20,2)

for i in k_options:
    classifier_pipeline = make_pipeline(preprocessing.StandardScaler(), KNeighborsClassifier(n_neighbors = i))
    scores = cross_val_score(classifier_pipeline, X_train, y_train, cv=10, scoring='accuracy')

    cv_scores.append(scores.mean())
    
MSE = [1 - x for x in cv_scores]

optimal_k = k_options[MSE.index(min(MSE))]
print("Melhor escolha de K é %d, com accuracy médio %f"%(optimal_k,max(cv_scores)))

plt.plot(k_options, MSE)
plt.xlabel('Number of Neighbors K')
plt.ylabel('Misclassification Error')
plt.show()
```

## Matriz de Confusão

```python
class_names = ['Ataque', 'Meio Campo', 'Defesa']

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
#         print("Normalized confusion matrix")
#     else:
#         print('Confusion matrix, without normalization')

#     print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

    
# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, y_pred)
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=class_names,
                      title='Confusion matrix, without normalization')

# Plot normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=class_names, normalize=True,
                      title='Normalized confusion matrix')

plt.show()
```
